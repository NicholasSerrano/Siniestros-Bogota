graficos plotly

```python
import plotly.express as px
import pandas as pd

# Suponiendo que tienes un DataFrame llamado df con las columnas 'fecha de vencimiento - Año' y 'CHOQUE'
# Puedes cargar tus datos desde un archivo o crear un DataFrame ficticio para este ejemplo
df = pd.DataFrame(df)

# Crea el gráfico de líneas con Plotly
fig = px.histogram(df, x= 'Marca', y= 'CHOQUE', title='Histograma: indice de siniestralidad por marca')
fig.update_layout(xaxis_title='Año de Vencimiento', yaxis_title='CHOQUE',)

# Muestra el gráfico
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponiendo que tu DataFrame se llama df
# Y que contiene las columnas 'CODIGO_ACCIDENTE', 'GRAVEDAD' y 'CODIGO_LOCALIDAD'

df = pd.DataFrame(df)

# Crear el gráfico de burbujas
plt.figure(figsize=(10, 6))
plt.scatter(df['CODIGO_ACCIDENTE'], df['GRAVEDAD'], s=df['CODIGO_LOCALIDAD']*10, alpha=0.5)

# Añadir etiquetas y título
plt.xlabel('Código de Accidente')
plt.ylabel('Gravedad')
plt.title('Gráfico de Burbujas')

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

```python
# Gráfico de dispersión con seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='fecha de vencimiento - Trimestre', y='CODIGO_LOCALIDAD', hue= 'CHOQUE', palette='viridis')
plt.title('Gráfico de Dispersión: Fecha de Vencimiento')
plt.xlabel('Fecha de Vencimiento - Trimestre')
plt.ylabel('Nombre Asegurado')
plt.grid(True)
plt.legend(title= 'CHOQUE')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponiendo que tu DataFrame se llama df y las columnas son 'CODIGO_ACCIDENTE' y 'fecha de vencimiento - Año'

# Leer el DataFrame
# df = pd.read_csv('tu_archivo.csv')  # Si los datos están en un archivo CSV

# Crear un DataFrame resumen que cuente los siniestros por año
siniestros_por_año = df.groupby('fecha de vencimiento - Año')['CODIGO_ACCIDENTE'].count()

# Crear el histograma de líneas
plt.figure(figsize=(10, 6))
siniestros_por_año.plot(kind='line', marker='o', color='b', linewidth=2)
plt.title('Cantidad de siniestros por año')
plt.xlabel('Año')
plt.ylabel('Cantidad de siniestros')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import plotly.graph_objs as go

# Suponiendo que tu DataFrame se llama df y las columnas son 'CODIGO_ACCIDENTE' y 'fecha de vencimiento - Año'

# Crear un DataFrame resumen que cuente los siniestros por año
siniestros_por_año = df.groupby('fecha de vencimiento - Año')['CODIGO_ACCIDENTE'].count().reset_index()

# Crear el gráfico de líneas
fig = go.Figure()

# Añadir la línea al gráfico
fig.add_trace(go.Scatter(x=siniestros_por_año['fecha de vencimiento - Año'], 
                         y=siniestros_por_año['CODIGO_ACCIDENTE'], 
                         mode='lines+markers', 
                         name='Siniestros por año'))

# Actualizar el diseño del gráfico
fig.update_layout(title='Cantidad de siniestros por año',
                   xaxis_title='Año',
                   yaxis_title='Cantidad de siniestros')

# Mostrar el gráfico
fig.show()
```
